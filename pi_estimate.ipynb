```python
import random
import matplotlib.pyplot as plt


def estimate_pi(n):
    num_points_circle = 0
    num_points_total = 0
    a = []
    b = []
    for _ in range(n):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        a.append(x)
        b.append(y)
        distance = x**2 + y**2
        if distance <= 1:
            num_points_circle += 1
            plt.scatter(a[_], b[_], color='r')
        else:
            plt.scatter(a[_], b[_], color='k')
        num_points_total += 1

    plt.show()


    return 4*num_points_circle/num_points_total


n = eval(input("Write the number of approximations you want to make: "))
print(estimate_pi(n))
```
